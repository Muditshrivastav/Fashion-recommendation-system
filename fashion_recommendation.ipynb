## Modify Preprocessing with Stemming/Lemmatization

### Subtask:
Update the existing text preprocessing pipeline to include a stemming or lemmatization step (e.g., using NLTK's Porter Stemmer or WordNet Lemmatizer) after lowercasing and punctuation removal, and apply this to the 'product_name' column.


**Reasoning**:
To begin updating the text preprocessing pipeline, I need to import the `nltk` library and ensure that the necessary 'punkt' and 'wordnet' data are downloaded. This is a prerequisite for using stemming or lemmatization functionalities.



```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
print("NLTK punkt and wordnet data downloaded.")
```

**Reasoning**:
Now that the necessary NLTK data is downloaded, I will import the `PorterStemmer` and define a function to apply stemming to the product names. This function will tokenize the text, stem each word, and then rejoin them, preparing the `product_name` column for improved TF-IDF vectorization.



```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def stem_text(text):
    words = text.split()
    stemmed_words = [stemmer.stem(word) for word in words]
    return ' '.join(stemmed_words)

men_data['product_name_stemmed'] = men_data['product_name'].apply(stem_text)
men_data[['product_name', 'product_name_stemmed']].head()
```

**Reasoning**:
Now that the product names have been stemmed, I will re-apply TF-IDF vectorization using the newly created `product_name_stemmed` column. This will create a new TF-IDF matrix that incorporates the stemmed terms, which is essential for the enhanced similarity search.



```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer_stemmed = TfidfVectorizer()
tfidf_matrix_stemmed = tfidf_vectorizer_stemmed.fit_transform(men_data['product_name_stemmed'])

print("Shape of TF-IDF matrix with stemmed names:", tfidf_matrix_stemmed.shape)
```

## Implement Updated Similarity Search Function

### Subtask:
Modify the `product_similarity_search` function to ensure that the input query also undergoes the same stemming/lemmatization preprocessing steps as the product names, allowing for consistent matching against the updated TF-IDF matrix.


**Reasoning**:
To implement the updated similarity search function, I need to define `product_similarity_search_stemmed` that preprocesses the query using the defined `stem_text` function, vectorizes it with `tfidf_vectorizer_stemmed`, calculates cosine similarity against `tfidf_matrix_stemmed`, and returns the top N similar products along with their original and stemmed names, image URLs, and similarity scores.



```python
from sklearn.metrics.pairwise import cosine_similarity

def product_similarity_search_stemmed(query, N=5):
    # Preprocess the query: lowercase, remove punctuation, and stem
    cleaned_query = query.lower()
    cleaned_query = re.sub(r'[^\w\s]', '', cleaned_query)
    stemmed_query = stem_text(cleaned_query)

    # Vectorize the cleaned and stemmed query using the stemmed TF-IDF vectorizer
    query_vector_stemmed = tfidf_vectorizer_stemmed.transform([stemmed_query])

    # Calculate cosine similarity between the query and all stemmed product names
    similarity_scores_stemmed = cosine_similarity(query_vector_stemmed, tfidf_matrix_stemmed).flatten()

    # Get the indices of the top N most similar products
    top_n_indices_stemmed = similarity_scores_stemmed.argsort()[-N:][::-1]

    # Retrieve the top N similar products and their details
    similar_products_stemmed = men_data.iloc[top_n_indices_stemmed][['product_name_stemmed', 'product_name', 'image_url']]
    similar_products_stemmed['similarity_score'] = similarity_scores_stemmed[top_n_indices_stemmed]

    return similar_products_stemmed

print("Product similarity search function (stemmed) defined.")
```

**Reasoning**:
Now that the `product_similarity_search_stemmed` function is defined, I will test it with the example query 'shirts' and display the results to demonstrate improved search relevance due to stemming.



```python
query = "shirts"
results_stemmed = product_similarity_search_stemmed(query, N=5)
print(f"Top 5 similar products for query '{query}' (with stemming):")
for index, row in results_stemmed.iterrows():
    print(f"Original Product: {row['product_name']}, Stemmed Product: {row['product_name_stemmed']}, Similarity: {row['similarity_score']:.2f}, Image URL: {row['image_url']}")
```

**Reasoning**:
Now that the `product_similarity_search_stemmed` function is working correctly with 'shirts', I will test it with a variety of other category queries ('shorts', 'trousers', 'shoes', 'perfumes', 'jackets') to demonstrate its improved relevance across different product types as required by the task.



```python
queries = ['shorts', 'trousers', 'shoes', 'jackets', 'hoodie', 'tshirt','blazers']

for query_term in queries:
    print(f"\n--- Searching for '{query_term}' (with stemming) ---")
    results_stemmed = product_similarity_search_stemmed(query_term, N=5)
    for index, row in results_stemmed.iterrows():
        print(f"Original Product: {row['product_name']}, Stemmed Product: {row['product_name_stemmed']}, Similarity: {row['similarity_score']:.2f}, Image URL: {row['image_url']}")
```

## Implement Recommendation Function

### Subtask:
Develop a Python function, let's call it `recommend_products`, that takes a product name as input. This function will internally use the previously defined `product_similarity_search_stemmed` function to find similar products based on the input product's name. It will return a list of recommended product names, their similarity scores, and their image URLs.


**Reasoning**:
To implement the `recommend_products` function, I will define it to take a product name and an optional number of recommendations, call the `product_similarity_search_stemmed` function internally, and then convert its DataFrame output into a list of dictionaries as specified.



```python
def recommend_products(product_name_input, N=5):
    # Use the existing product_similarity_search_stemmed function
    recommended_df = product_similarity_search_stemmed(product_name_input, N=N)

    # Convert the DataFrame to a list of dictionaries
    recommendations_list = recommended_df.apply(lambda row: {
        'product_name': row['product_name'],
        'image_url': row['image_url'],
        'similarity_score': row['similarity_score']
    }, axis=1).tolist()

    return recommendations_list

print("Recommendation function 'recommend_products' defined.")
```

```python
import joblib
import pandas as pd
import re
from sklearn.metrics.pairwise import cosine_similarity

# 1. Load the saved components
loaded_vectorizer = joblib.load('tfidf_vectorizer.joblib')
loaded_matrix = joblib.load('tfidf_matrix.joblib')
loaded_men_data = pd.read_csv('cleaned_men_data.csv')

# 2. Define the search logic using loaded components
def verify_similarity_search(query, N=5):
    # Preprocess query exactly as before
    from nltk.stem import PorterStemmer
    stemmer = PorterStemmer()

    cleaned_query = query.lower()
    cleaned_query = re.sub(r'[^\w\s]', '', cleaned_query)
    stemmed_query = ' '.join([stemmer.stem(word) for word in cleaned_query.split()])

    # Vectorize and calculate similarity
    query_vec = loaded_vectorizer.transform([stemmed_query])
    scores = cosine_similarity(query_vec, loaded_matrix).flatten()

    top_indices = scores.argsort()[-N:][::-1]
    results = loaded_men_data.iloc[top_indices].copy()
    results['similarity_score'] = scores[top_indices]

    return results[['product_name', 'similarity_score', 'image_url']]

# 3. Perform test query
test_query = 'trousers'
verification_results = verify_similarity_search(test_query)

print(f'Verification results for query: {test_query}')
print(verification_results.to_string(index=False))
```
